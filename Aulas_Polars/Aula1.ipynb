prompt: 
Neste dataset temos:
- Colunas com valores numéricos:
"Age, Experience, Income, CCAvg, Mortgage"
- Colunas com dados categóricos:
"Family, Education, Personal Loan, Securities Account, CD Account, Online e CreditCard"
Crie códigos em Python com a biblioteca Seaborn para visualizar
a distribuição dos dados numéricos e para visualizar as quantidades dos dados categóricos.
Quero visualizar os gráficos em relação à variável "Personal Loan",
usando cores diferentes para as categorias "Não Aceitou" e "Aceitou".
- Use somente a biblioteca polars
- O dataset ja esta na variavel df

````python
import polars as pl
import matplotlib.pyplot as plt

# Numeric variables
numeric_vars = ['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']

# Categorical variables
categorical_vars = ['Family', 'Education', 'Securities Account', 'CD Account', 'Online', 'CreditCard']

# Plotting numeric variables
fig, axes = plt.subplots(len(numeric_vars), 1, figsize=(10, 5 * len(numeric_vars)))

for i, var in enumerate(numeric_vars):
    # Group by 'Personal Loan' and calculate mean and median
    grouped = df.group_by("Personal Loan").agg([
        pl.col(var).mean().alias(f"{var}_mean"),
        pl.col(var).median().alias(f"{var}_median")
    ]).sort("Personal Loan")

    # Plotting means
    axes[i].bar(grouped['Personal Loan'].cast(pl.Utf8), grouped[f'{var}_mean'], label=f'{var} Mean')
    
    # Plotting medians
    axes[i].bar(grouped['Personal Loan'].cast(pl.Utf8), grouped[f'{var}_median'], label=f'{var} Median')
    
    axes[i].set_title(f'{var} by Personal Loan Status')
    axes[i].set_xlabel('Personal Loan Status (0: Not Accepted, 1: Accepted)')
    axes[i].set_ylabel(var)
    axes[i].legend()

plt.tight_layout()
plt.show()

# Plotting categorical variables
fig, axes = plt.subplots(len(categorical_vars), 1, figsize=(10, 5 * len(categorical_vars)))

for i, var in enumerate(categorical_vars):
    # Group by 'Personal Loan' and count the occurrences of each category
    counts = df.group_by(["Personal Loan", var]).agg(pl.count().alias("count")).sort(["Personal Loan", var])
    
    # Get unique values of the categorical variable
    categories = df.select(pl.col(var)).unique().sort(var).to_series().to_list()
    
    # Create a dictionary to store the counts for each category and Personal Loan status
    data = {}
    for category in categories:
        data[category] = []
        for loan_status in [0, 1]:
            count = counts.filter((pl.col("Personal Loan") == loan_status) & (pl.col(var) == category))['count'].to_list()
            if count:
                data[category].append(count[0])
            else:
                data[category].append(0)
    
    # Plotting the counts
    width = 0.35  # the width of the bars
    x = range(2)  # the x locations for the groups
    
    for j, category in enumerate(categories):
        offset = width * j
        axes[i].bar([pos + offset for pos in x], data[category], width, label=str(category))
    
    axes[i].set_title(f'{var} by Personal Loan Status')
    axes[i].set_xlabel('Personal Loan Status (0: Not Accepted, 1: Accepted)')
    axes[i].set_ylabel('Count')
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(['0', '1'])
    axes[i].legend(title=var)

plt.tight_layout()
plt.show()
````
